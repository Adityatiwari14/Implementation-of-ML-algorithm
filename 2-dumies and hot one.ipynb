```python
# https://github.com/codebasics/py/blob/master/ML/5_one_hot_encoding/one_hot_encoding.ipynb


import numpy as np

import pandas as pd
df = pd.read_csv(r"C:\Users\adity\OneDrive\Desktop\ML\homeprices.csv")
```

```python
dummies = pd.get_dummies(df['town'])
dummies=dummies.astype(int)
```

```python
dummies.head()
```

```python
merged = pd.concat([df,dummies],axis='columns')
```

```python
merged
```

```python
final = merged.drop(['town'], axis='columns')
final
```

```python
final = final.drop(['west windsor'],axis ='columns') #dummy problem does not occur
```

```python
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop(['price'],axis = 'columns')
```

```python
X
```

```python
y= final['price']
```

```python
y
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]])
# in order of your x
```

```python
model.predict([[3800,0,0]])
```

```python
model.score(X,y)
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
```

```python
dfle=df
dfle['town'] = le.fit_transform(dfle.town)
```

```python
dfle
```

```python
X=df[['town','area']].values #.values to get two dimentional array and not dataframe
X
```

```python
y=dfle.price
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')

```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]
X
```

```python
model.fit(X,y)
```

```python
model.predict([[0,1,3400]]) # 3400 sqr ft home in west windsor
```

```python
model.predict([[1,0,2800]]) # 2800 sqr ft home in robbinsville
```
