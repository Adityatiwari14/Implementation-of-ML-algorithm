```python
df= pd.read_csv(r"C:\Users\adity\OneDrive\Desktop\ML\titanic.csv")
```

```python
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
target = df.Survived
```

```python
input = df.drop('Survived', axis='columns')
```

```python
dummies = pd.get_dummies(input.Sex).astype(int)
dummies.head(3)
```

```python
input= pd.concat([input , dummies],axis ='columns')
input.head()
```

```python
input=input.drop('Sex',axis='columns')
```

```python
input.head()
```

```python
input.isna().sum()
```

```python
input['Age']=input['Age'].fillna(input.Age.mean())
```

```python
input['Fare']=input['Fare'].fillna(input.Fare.mean())
```

```python
input.head(10)
```

```python
input=input.drop('male',axis='columns')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(input,target,test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train , y_train)
```

```python
model.score(X_test , y_test)
```
